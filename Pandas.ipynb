13) We can also find out the no of data satisfying the condition by just using the shape attribute


```
df[mask].shape
We know this will return a tuple (row,col) here row will be the desired count.

// Better way of doing this
df[mask].shape[0] will return row only (as we know shape return a tuple we can access it like this)
```



````python
def EnterCityName(city):
    mask = df['city'] == city
    return df[mask].shape[0];
````

````python
EnterCityName('Kolkata')
````
